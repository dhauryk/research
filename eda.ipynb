```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from transformers import BertTokenizer, TFBertForMaskedLM, BertConfig
from sklearn.model_selection import train_test_split

# Load Santander Product Recommendation dataset
data_path = "santander-product-recommendation.csv"  # Update with actual path
df = pd.read_csv(data_path)  # Read dataset into a pandas DataFrame

# Preprocess data
def preprocess_data(df):
    df = df.sort_values(['ncodpers', 'fecha_dato'])  # Sort data by customer ID and date
    df['product_sequence'] = df.groupby('ncodpers')['producto'].apply(lambda x: ' '.join(x))  # Create sequences of products per customer
    unique_sequences = df[['ncodpers', 'product_sequence']].drop_duplicates()  # Remove duplicate sequences per customer
    return unique_sequences

data = preprocess_data(df)  # Apply preprocessing function

# Split data into training and testing sets
train_data, test_data = train_test_split(data, test_size=0.1, random_state=42)

# Load pre-trained BERT tokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# Tokenize product sequences for training and testing
train_encodings = tokenizer(list(train_data['product_sequence']), padding=True, truncation=True, return_tensors="tf")
test_encodings = tokenizer(list(test_data['product_sequence']), padding=True, truncation=True, return_tensors="tf")

# Convert tokenized data into TensorFlow datasets for efficient processing
train_dataset = tf.data.Dataset.from_tensor_slices((dict(train_encodings), train_encodings['input_ids']))
test_dataset = tf.data.Dataset.from_tensor_slices((dict(test_encodings), test_encodings['input_ids']))
train_dataset = train_dataset.shuffle(1000).batch(8).prefetch(tf.data.AUTOTUNE)  # Shuffle and batch training data
test_dataset = test_dataset.batch(8).prefetch(tf.data.AUTOTUNE)  # Batch test data

# Define BERT4Rec model configuration
config = BertConfig.from_pretrained("bert-base-uncased", num_hidden_layers=4, num_attention_heads=4, hidden_size=256)
model = TFBertForMaskedLM(config)  # Create BERT-based model for Masked Language Modeling

# Compile model with optimizer and loss function
optimizer = tf.keras.optimizers.Adam(learning_rate=5e-5)  # Adam optimizer with learning rate 5e-5
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)  # Loss function for masked language modeling
tmodel.compile(optimizer=optimizer, loss=loss)

# Train model and store history for analysis
history = model.fit(train_dataset, validation_data=test_dataset, epochs=3)

# Visualization of training loss
tplt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Training and Validation Loss Over Epochs')
plt.show()

# Make predictions on test data
predictions = model.predict(test_encodings['input_ids']).logits  # Get model logits (raw scores)
predicted_tokens = np.argsort(-predictions, axis=-1)[:, :, :7]  # Extract top 7 predicted tokens per sequence

# Compute MAP@7 metric
def map_at_7(preds, actuals):
    avg_precisions = []  # List to store average precisions per sequence
    for pred, actual in zip(preds, actuals):
        hits = 0  # Track number of correct predictions
        sum_precisions = 0  # Sum of precisions at each rank
        for i, p in enumerate(pred[:7]):  # Iterate over top 7 predictions
            if p in actual:  # Check if prediction is in actual product list
                hits += 1  # Increase hit count
                sum_precisions += hits / (i + 1)  # Compute precision at i-th rank
        avg_precisions.append(sum_precisions / min(7, len(actual)))  # Compute average precision
    return np.mean(avg_precisions)  # Return mean of all average precisions

# Convert actual product sequences to token IDs
actual_tokens = [tokenizer.convert_tokens_to_ids(seq.split()) for seq in test_data['product_sequence']]
map7_score = map_at_7(predicted_tokens, actual_tokens)  # Compute MAP@7 score
print(f"MAP@7 Score: {map7_score:.4f}")  # Print final MAP@7 score

# Save trained model and tokenizer for future use
model.save_pretrained("bert4rec_santander")
tokenizer.save_pretrained("bert4rec_santander")

print("Model training completed and saved.")
```
