```python
import cv2
import math
import cvzone
from ultralytics import YOLO
import os


yolo_model = YOLO("/home/dhauryk/!DUNAI/runs/detect/train/weights/best.pt")
```

```python
class_labels = ['Bodypanel-Dent', 'Front-Windscreen-Damage', 'Headlight-Damage', 
'Rear-windscreen-Damage', 'RunningBoard-Dent', 'Sidemirror-Damage', 'Signlight-Damage', 
'Taillight-Damage', 'bonnet-dent', 'boot-dent', 'doorouter-dent', 'fender-dent', 
'front-bumper-dent', 'pillar-dent', 'quaterpanel-dent', 'rear-bumper-dent', 'roof-dent']
```

```python

from utils.utils_files import get_all_image_file_names_in_folder
from rembg import remove 

all_paths = get_all_image_file_names_in_folder("microchip")

for image_path in all_paths:
    img = cv2.imread(image_path, cv2.IMREAD_COLOR)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # img = remove(img)
    # img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = yolo_model(img)

    for r in results:
        boxes = r.boxes
        for box in boxes:
            x1, y1, x2, y2 = box.xyxy[0]
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)

            w, h = x2 - x1, y2 - y1
            
            conf = math.ceil((box.conf[0] * 100)) / 100
            cls = int(box.cls[0])

            if conf > 0.1:
                cvzone.cornerRect(img, (x1, y1, w, h), t=7)
                cvzone.putTextRect(img, f'{class_labels[cls]} {conf}', (x1, y1 - 10), scale=0.8, thickness=1, colorR=(255, 0, 0))

    cv2.imwrite(os.path.join("output", "car_damage",  f"{''.join(os.path.basename(image_path).split('.')[:-1])}.jpeg"), 
                        cv2.cvtColor(img, cv2.COLOR_RGB2BGR))
```

```python
# Display the image with detections
cv2.imshow("Image", img)

# Close window when 'q' button is pressed
while True:
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
cv2.waitKey(1)
```
